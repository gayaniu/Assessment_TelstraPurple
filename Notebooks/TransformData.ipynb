```
spark.conf.set(
  "fs.azure.account.key.telstrapurpledatastore.dfs.core.windows.net",
  dbutils.secrets.get(scope="spark_conf_set_key",key="dtabaricksAccessKey-context")
)
```

```
dbutils.fs.ls("abfss://datalakestore@telstrapurpledatastore.dfs.core.windows.net/Raw")
#set the data lake file location:
file_location = "abfss://datalakestore@telstrapurpledatastore.dfs.core.windows.net/Raw/Feed_data.json"
 
#read the data to dataframe df

df = spark.read.format('org.apache.spark.sql.json') \
        .load(file_location) 
#display(df)
#df.printSchema()

from pyspark.sql.functions import col, explode

transformed_df = df.withColumn("items",explode("rss.channel.item")).select(col("items.address").alias("address"),
        col("items.brand").alias("brand"),
        col("items.date").alias("date"),
        col("items.description").alias("description"),
        col("items.latitude").alias("latitude"),
        col("items.location").alias("location"),
        col("items.longitude").alias("longitude"),
        col("items.phone").alias("phone"),
        col("items.price").alias("price") ,
        col("items.site-features").alias("sitefeatures"),
        col("items.title").alias("title"),
        col("items.trading-name").alias("tradingname")       
       )

transformed_df.show()

```

```
#writing the transformed data to the Parquet format and store in the Datalake storage RAW/FuelWatch container
transformed_df.write.parquet("abfss://datalakestore@telstrapurpledatastore.dfs.core.windows.net/Raw/FuelWatch/feed.parquet", mode='overwrite')
```

```
#preparing the sufix for the file to dave to destination. 
from datetime import datetime, date, time
now =datetime.now()
date_string = now.strftime('%Y-%m-%d-%H:%M:%S.%f')
```

```
#Moving the processed file to processed folder. 
dbutils.fs.mv('abfss://datalakestore@telstrapurpledatastore.dfs.core.windows.net/Raw/Feed_data.json',     'abfss://datalakestore@telstrapurpledatastore.dfs.core.windows.net/Raw/Processed/Feed_data_Processed_'+date_string+'.json')
```
